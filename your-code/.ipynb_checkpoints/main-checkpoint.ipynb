```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn.cluster import KMeans
from sklearn import metrics

model = KMeans(n_clusters=8,random_state=42)
model.fit(X)

score = metrics.silhouette_score(X, model.labels_, metric='euclidean', 
                                   sample_size=len(X))

print('Silhouette score with 8 clusers: ',score)
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
model = KMeans(n_clusters=5,random_state=42)
model.fit(X)

score = metrics.silhouette_score(X, model.labels_, metric='euclidean', 
                                   sample_size=len(X))

print('Silhouette score with 5 clusers: ',score)
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
model = KMeans(n_clusters=3,random_state=42)
model.fit(X)

score = metrics.silhouette_score(X, model.labels_, metric='euclidean', 
                                   sample_size=len(X))

print('Silhouette score with 3 clusers: ',score)
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from yellowbrick.cluster import KElbowVisualizer

# Generate synthetic dataset with 8 random clusters
#X, y = make_blobs(n_samples=1000, n_features=12, centers=8, random_state=42)

# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model,metric='distortion', k=(2,12))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show() 
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from yellowbrick.cluster import KElbowVisualizer


model = KMeans()
visualizer = KElbowVisualizer(model,metric='silhouette', k=(2,12))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show() 

# La linea gris es el tiempo de entrenamiento
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model,metric='calinski_harabasz', k=(2,12))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show() 
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer
#from yellowbrick.cluster import calin
model = KMeans(4, random_state=42)
visualizer = SilhouetteVisualizer(model, colors='yellowbrick')

visualizer.fit(X)       
visualizer.show()  
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
from sklearn.manifold import TSNE
fig, ax = plt.subplots(1, 4, figsize=(15, 7))
# Generamos Perplejidades
# OBTENER LA MS CERTERA n^(1/2)
for i, perp in enumerate([5,12, 30, 50]):
  # Usamos TSNE
    tsne = TSNE(perplexity=perp,random_state=123)
    # Vamos uniendo cada perpeplijdad segpun el componente
    x_embedded = tsne.fit_transform(X)
    # De todos modos, todo esto es una forma larga de decir que los ejes de una 
    # grfica t-SNE no son fcilmente interpretables en trminos de los ejes / 
    # unidades de los datos originales de alta dimensin. Por tanto, no tratara 
    # de interpretar cuantitativamente una grfica t-SNE; t-SNE es solo una tcnica de visualizacin, nada ms.
    # https://stats.stackexchange.com/questions/254090/what-are-the-axes-of-a-t-sne-scatterplot
    
    ax[i].scatter(x_embedded[:, 0], x_embedded[:, 1], c=model.labels_,cmap='rainbow')
    ax[i].set_title("Perplexity = {}".format(perp))
    
plt.show()
```
